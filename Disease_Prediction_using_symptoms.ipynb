Remove the spaces between names

```python
def remove_space_between_word(dataset):
    for col in dataset.columns:
        for i in range(len(dataset[col])):
            if (type(dataset[col][i]) == str ):
                dataset[col][i] = dataset[col][i].strip()
                dataset[col][i] = dataset[col][i].replace(" ", "_")
    return df
```

# **One_Hot_Encoding**

create dataset with only zero values

```python
d = pd.DataFrame(0, index=np.arange(len(df)), columns=unique_symptoms)
d
```

encoding the above dataset(replacing zeros with one when there is match)

```python
def one_hot_encode(df):
  for d1 in range(0,len(df)):
    #print(d1)
    for i in d.columns:
      #print(i)
      for x in range(0,17):
        find = df.iloc[d1, x]
        #print(find)
        if i == find:
          d.iloc[d1, d.columns.get_loc(i)] = 1

  return d
```

```python
df = one_hot_encode(df)
```

```python
all_disease = df['Disease']
all_disease = list(all_disease)
```
